```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set the visual style for plots
sns.set(style="whitegrid")
```

```python
# Load the dataset
df = pd.read_csv('heart.csv')

# Display basic information
print("Dataset Shape:", df.shape)
print("\nFirst 5 rows:")
display(df.head())

# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())
```

# initial number of rows and columns in the dataset before any cleaning was performed (1025, 14).

```python
# Check for duplicates before cleaning
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows found: {duplicates}")

# Remove duplicates
df = df.drop_duplicates()

# Final check after cleaning
print(f"New dataset shape after cleaning: {df.shape}")
```

# After using df.drop_duplicates(), we can see that the duplicates were successfully removed new Shape: (302, 14).


```python
plt.figure(figsize=(6, 4))
sns.countplot(x='target', data=df, palette='viridis')
plt.title('Heart Disease Presence (0 = No, 1 = Yes)')
plt.show()
# Insight: The dataset is well-balanced between patients with and without heart disease.
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['age'], kde=True, color='skyblue')
plt.title('Age Distribution of Patients')
plt.show()
# Insight: Most patients in this study are middle-aged, primarily between 50 and 60 years old.
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='sex', hue='target', data=df, palette='Set2')
plt.title('Heart Disease Frequency for Sex')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.show()
# Insight: In this specific dataset, male patients show a higher frequency of heart disease.
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='RdBu', fmt='.2f')
plt.title('Correlation Heatmap of Features')
plt.show()
# Insight: Features like 'cp' and 'thalach' have the strongest correlation with the target.
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='chol', hue='target', data=df, alpha=0.7)
plt.title('Cholesterol vs Age')
plt.show()
# Insight: There is no simple linear separation between age and cholesterol for heart disease.
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='cp', hue='target', data=df)
plt.title('Chest Pain Type vs Target')
plt.show()
# Insight: Non-anginal chest pain (Type 2) is highly associated with positive heart disease cases.
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='target', y='thalach', data=df)
plt.title('Maximum Heart Rate vs Heart Disease')
plt.show()
# Insight: Patients with heart disease tend to reach a higher maximum heart rate (thalach).
```
